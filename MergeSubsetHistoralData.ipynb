```python
import pandas
import pygrib
import pickle
import os.path
import time
import subprocess
```

```python

def merge_subset_grib_files(grib_dir, out_file, year, south_lat, north_lat, west_lng, east_lng):

    saved_varibles = [
        'u-component of wind',
        'v-component of wind',
    ]
    
    # if os.path.exists(out_file):
    #     print(f"error {out_file} exists")
    #     return

    with open(out_file, "wb") as out_f:
        #for day in range(1,25):
        #for day in range(30,32):
        for day in range(1,20):
            for fctime in [0,6,12,18]:
                #for simulation_time in range(0,6):
                for simulation_time in [0]:
                    tic = time.time()
                    grib_file = f"{grib_dir}/{year}/{year}07{day:02}-{fctime:02}-gfs.t{fctime:02}z.pgrb2.0p25.f{simulation_time:03}"
                    print(f"{grib_file} >> {out_file}")
                    if not os.path.exists(grib_file):
                        print("grib file not found, exiting")
                        return
                    grbs = pygrib.open(grib_file)
                    for grb in grbs:
                        try:
                            # Extract the variable name
                            variable_name = grb.parameterName  

                            if variable_name in saved_varibles and grb.level == 10:
                                # Extract the data within the bounding box
                                data, lats, lons = grb.data(south_lat, north_lat, west_lng, east_lng)
                                
                                if data is not None:
                                    #print(f"    Saving: '{variable_name}' level={grb.level}")
                                    out_f.write(grb.tostring())  # Write the subset to the output file
                                #else:
                                #    print(f"    '{variable_name}' level={grb.level} is empty")
                            #else:
                            #    print(f"    Skipping: '{variable_name}' level={grb.level}")
                            
                        except Exception as e:
                            print(f"    Error extracting '{variable_name}' in {grib_file} due to error: {e}")

                    print(f"    {time.time()-tic:.1f}s")
                
    print("Subset merging complete.")
```

```python
min_lat = 21.1
max_lat = 34.1
max_lng = -118
min_lng = -158
```

```python
#grib_dir ='historical_gribs'
grib_dir = '/Volumes/Gribs'
#grib_dir = 'historical_gribs'
year = 2021
out_file = f'/Volumes/Gribs/MergedSubsetJuly{year}.grib'

merge_subset_grib_files(grib_dir, out_file, year, min_lat, max_lat, max_lng, min_lng)
```
